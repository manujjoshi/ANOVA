# `Shapiro wilk test`

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

- It is used find the normality of the distribution
- It is used only for Univariant Continuous Data

##### H0: Null Hypothesis:Sample is drawn from a Normally Distributed Population
##### Ha: Alternate Hypothesis:Sample is drawn from a Not Normally Distributed Population

```python
x_normal = np.random.normal(100,400,50000)  #generating random normal values
```

```python
sns.distplot(x_normal)
```

```python
result = stats.shapiro(x_normal)    
result
```

### Here p-value > 0.05 So, Accept Null Hypothesis.
### So,It is a NormalDistribution.

```python
x = np.random.randint(100,400,50000)  #generating random values
```

```python
sns.distplot(x)
```

```python
result = stats.shapiro(x)    
result
```

### Here p-value < 0.05 So, Reject Null Hypothesis.
### So,It is not a NormalDistribution.
