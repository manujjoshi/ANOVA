## 2. `Stratified sample` 

```python
st_sample = data.groupby("Geography").sample(5)
st_sample
```

```python
st_sample["Geography"].value_counts()
```

## 3. `Cluster sampling`

```python
country = data["Geography"].unique().tolist()
country
```

```python
import random
random.choice(country)
```

```python
random.sample(country,2)
```

```python
data.loc[data["Geography"].isin(['Spain', 'France'])] # using isin 
```

## 5. `Multistage Stratified sampling`

```python
st_sample = data.groupby("Geography").sample(500)
st_sample
```

```python
mul_st = st_sample.groupby(["Geography","Gender"]).sample(100)
mul_st
```

```python
mul_st['Geography'].value_counts()
```

```python
mul_st['Gender'].value_counts()
```
